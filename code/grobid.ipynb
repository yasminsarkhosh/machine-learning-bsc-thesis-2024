```python
import os
import pandas as pd
from lxml import etree

def parse_xml_and_extract_headers(file_path):
    # Parse the XML file
    tree = etree.parse(file_path)
    root = tree.getroot()

    # Define a namespace dictionary to use with XPath
    ns = {'tei': 'http://www.tei-c.org/ns/1.0'}

    # Extract headers and their texts
    headers = root.xpath('//tei:head', namespaces=ns)
    
    data = []  # Initialize a list to store extracted data
    
    for header in headers:
        text = ' '.join([p.text for p in header.getparent().findall('tei:p', ns) if p.text])
        data.append({
            'Header Number': header.get('n'),
            'Header Title': header.text,
            'Text': text
        })

    # Create a DataFrame from the extracted data
    df = pd.DataFrame(data, columns=['Header Number', 'Header Title', 'Text'])
    return df

def process_xml_folder(folder_path):
    all_data_frames = []  # List to hold data frames from all files

    # Loop through each file in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".xml"):  # Process only XML files
            file_path = os.path.join(folder_path, file_name)
            df = parse_xml_and_extract_headers(file_path)  # Parse the file and extract headers
            all_data_frames.append(df)

    # Concatenate all DataFrames into one
    if all_data_frames:
        final_df = pd.concat(all_data_frames, ignore_index=True)
    else:
        final_df = pd.DataFrame()  # Return an empty DataFrame if no XML files were processed

    return final_df

# Folder path - Update this to the path where your XML files are stored

# Process the folder and create a DataFrame with all headers
df_headers = process_xml_folder(folder_path)
folder_path = '/Users/yasminsarkhosh/Documents/GitHub/machine-learning-bsc-thesis-2024/code/vol10/'
df_headers.to_csv("/Users/yasminsarkhosh/Documents/GitHub/machine-learning-bsc-thesis-2024/code/vol10_headers.csv", index=False)
```
